```python
# ============================================================================
# COMBINE FINAL METRICS FROM ALL EXPERIMENT FOLDERS SUMMARIZATION WITHOUT RAG
# ============================================================================
import pandas as pd
from pathlib import Path
from google.colab import drive

drive.mount('/content/drive')

PROJECT_DIR = Path("/content/drive/MyDrive/Colab Notebooks/Masters/Masters_Thesis_Project")

EXPERIMENTS_DIR = PROJECT_DIR / "outputs" / "experiments"

OUTPUT_DIR = PROJECT_DIR / "outputs" / "final_comparison" / "summarization" / "non_rag"
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

COMBINED_FILE = OUTPUT_DIR / "non_rag_summarization_final_metrics_all_experiments.csv"

all_rows = []

# Loop through each experiment folder
for exp_folder in EXPERIMENTS_DIR.iterdir():

    if exp_folder.is_dir():

        # Find final metrics file inside this folder
        final_metric_files = list(exp_folder.glob("*_final_metrics.csv"))

        if len(final_metric_files) == 0:
            print(f"No final metrics found in: {exp_folder.name}")
            continue

        # There should be only one
        file = final_metric_files[0]

        try:
            temp_df = pd.read_csv(file)

            if len(temp_df) == 0:
                print(f"Empty file: {file}")
                continue

            row = temp_df.iloc[0].to_dict()

            # Optional: add folder name for safety/debugging
            row["experiment_folder"] = exp_folder.name

            all_rows.append(row)

            print(f"Loaded: {exp_folder.name}")

        except Exception as e:
            print(f"Failed reading {file}: {e}")

# Combine all
combined_df = pd.DataFrame(all_rows)

# Save
combined_df.to_csv(COMBINED_FILE, index=False)

print("\n==============================")
print("FINAL COMBINED FILE SAVED AT:")
print(COMBINED_FILE)
print("==============================")

display(combined_df)
```
