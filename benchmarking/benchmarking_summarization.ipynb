# Benchmarking and Comparative Analysis of Prompt-Based Summarization Experiments

## Overview

This notebook performs benchmarking and comparative evaluation of prompt-based job description summarization experiments conducted on the Djinni Job Descriptions Dataset.

The experiments were conducted using:

- Multiple prompting techniques
- Multiple Large Language Models (LLMs)
- Retrieval-Augmented Generation (RAG) and Non-RAG configurations

The objective of this benchmarking pipeline is to:

1. Evaluate summarization quality across experiments
2. Rank prompting techniques for each LLM
3. Rank LLMs for each prompting technique
4. Compare RAG and Non-RAG performance
5. Identify the best-performing experimental configurations
6. Generate quantitative benchmark statistics and visualizations

---

# Experimental Configurations

## Prompting Techniques Evaluated

- Zero-Shot Prompting
- Few-Shot Prompting
- Chain-of-Thought Prompting
- Style-Based Prompting
- Self-Consistency Prompting

---

## Large Language Models Evaluated

- llama-3.1-8b-instant
- llama-3.3-70b-versatile
- openai/gpt-oss-120b

---

## Experimental Variants

- Summarization without RAG
- Summarization with RAG

---

# Benchmarking Methodology

The benchmarking pipeline performs:

- Metric normalization using Min-Max Scaling
- Composite benchmark score calculation using weighted metrics
- Prompt ranking analysis
- Model ranking analysis
- Overall experiment ranking
- RAG vs Non-RAG comparison
- Benchmark visualization generation

---

# Evaluation Metrics Used

## Positive Metrics (Higher is Better)

- ROUGE-L F1 Score
- Semantic Similarity
- Information Retention Ratio (IRR)
- Top-K Skill Coverage
- Global Skill Overlap

---

## Negative Metrics (Lower is Better)

- Hallucination Rate
- Hallucination Density
- Hallucination Prevalence
- Compression Ratio

This notebook generates:

- Prompt benchmarking tables

- Benchmark score visualizations
- Benchmark CSV files for thesis reporting

# Expected Outcome

The benchmarking results generated in this notebook support the comparative analysis chapter of the thesis and provide quantitative evidence for evaluating the effectiveness of prompting strategies and Retrieval-Augmented Generation in job description summarization tasks.

```python
# =====================================================================
# BENCHMARKING PIPELINE FOR SUMMARIZATION EXPERIMENTS
# =====================================================================
# Purpose:
# This notebook performs benchmarking and ranking of summarization
# experiments conducted with:
#
# 1. Multiple prompting techniques
# 2. Multiple LLMs
# 3. RAG and Non-RAG configurations
#
# The notebook performs:
#
# - Metric normalization
# - Composite benchmark score calculation
# - Prompt ranking
# - Model ranking
# - Overall experiment ranking
# - RAG vs Non-RAG comparison
# - Benchmark visualizations
#
# =====================================================================

# =====================================================================
# 1. IMPORT REQUIRED LIBRARIES
# =====================================================================

import pandas as pd
import numpy as np

from pathlib import Path

from sklearn.preprocessing import MinMaxScaler

import matplotlib.pyplot as plt

# =====================================================================
# 2. DISPLAY INITIAL LOGS
# =====================================================================

print("=" * 70)
print("SUMMARIZATION BENCHMARKING PIPELINE")
print("=" * 70)

# =====================================================================
# 3. DEFINE INPUT FILE PATHS
# =====================================================================
# Update these paths if necessary.
# These files contain final metrics from:
#
# - Non-RAG summarization experiments
# - RAG summarization experiments
#
# =====================================================================

# =====================================================================
# MOUNT GOOGLE DRIVE
# =====================================================================

from google.colab import drive

drive.mount('/content/drive')

print("Google Drive mounted successfully")

# =====================================================================
# DEFINE PROJECT DIRECTORY
# =====================================================================
# Update this path according to your project folder location
#
# Example:
# /content/drive/MyDrive/Colab Notebooks/Masters/LJMU-masters-thesis
#
# =====================================================================

from pathlib import Path

PROJECT_DIR = Path(
    "/content/drive/MyDrive/Colab Notebooks/Masters/LJMU-masters-thesis"
)

print("\nProject directory:")
print(PROJECT_DIR)

# =====================================================================
# DEFINE INPUT FILE PATHS
# =====================================================================
# These files should contain:
#
# - Final metrics for Non-RAG experiments
# - Final metrics for RAG experiments
#
# Example structure:
#
# outputs/
#   final_comparison/
#       summarization/
#           non_rag/
#           rag/
#
# =====================================================================

NON_RAG_FILE = (
    PROJECT_DIR
    / "outputs"
    / "final_comparison"
    / "summarization"
    / "non_rag"
    / "non_rag_final_metrics.csv"
)

RAG_FILE = (
    PROJECT_DIR
    / "outputs"
    / "final_comparison"
    / "summarization"
    / "rag"
    / "rag_final_metrics.csv"
)

print("\nInput files:")

print("\nNon-RAG file:")
print(NON_RAG_FILE)

print("\nRAG file:")
print(RAG_FILE)

# =====================================================================
# CREATE OUTPUT DIRECTORY
# =====================================================================
# All benchmarking outputs will be saved here.
#
# =====================================================================

OUTPUT_DIR = (
    PROJECT_DIR
    / "outputs"
    / "benchmarking"
    / "summarization"
)

OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

print("\nOutput directory created successfully")
print(OUTPUT_DIR)

# =====================================================================
# 5. LOAD EXPERIMENT RESULT FILES
# =====================================================================

non_rag_df = pd.read_csv(NON_RAG_FILE)

rag_df = pd.read_csv(RAG_FILE)

print("\nFiles loaded successfully")

print("\nNon-RAG dataframe shape:", non_rag_df.shape)

print("RAG dataframe shape    :", rag_df.shape)

# =====================================================================
# 6. ADD RAG TYPE LABEL
# =====================================================================
# This column helps differentiate:
#
# - RAG experiments
# - Non-RAG experiments
#
# =====================================================================

non_rag_df["rag_type"] = "NON_RAG"

rag_df["rag_type"] = "RAG"

# =====================================================================
# 7. CLEAN PROMPT TYPE NAMES
# =====================================================================
# Some experiment outputs may contain:
#
# - _with_rag
# - _without_rag
#
# These suffixes are removed to standardize prompt names.
#
# =====================================================================

rag_df["prompt_type"] = (
    rag_df["prompt_type"]
    .str.replace("_with_rag", "", regex=False)
    .str.replace("with_rag", "", regex=False)
)

non_rag_df["prompt_type"] = (
    non_rag_df["prompt_type"]
    .str.replace("_without_rag", "", regex=False)
    .str.replace("without_rag", "", regex=False)
)

print("\nPrompt types after cleaning")

print("\nNon-RAG prompt types:")
print(non_rag_df["prompt_type"].unique())

print("\nRAG prompt types:")
print(rag_df["prompt_type"].unique())

# =====================================================================
# 8. COMBINE RAG AND NON-RAG DATAFRAMES
# =====================================================================
# This creates a unified dataframe for benchmarking.
#
# =====================================================================

combined_df = pd.concat(
    [non_rag_df, rag_df],
    ignore_index=True,
    sort=False
)

print("\nCombined dataframe shape:", combined_df.shape)

# Save combined dataframe
combined_df.to_csv(
    OUTPUT_DIR / "combined_rag_nonrag_metrics.csv",
    index=False
)

print("\nCombined dataframe saved successfully")

# =====================================================================
# 9. DEFINE BENCHMARK METRICS
# =====================================================================
# Positive metrics:
# Higher values indicate better performance.
#
# Negative metrics:
# Lower values indicate better performance.
#
# =====================================================================

positive_metrics = [

    "rougeL_f1_mean",

    "semantic_similarity_mean",

    "irr_mean",

    "top_k_skill_coverage_mean",

    "global_skill_overlap"
]

negative_metrics = [

    "hallucination_rate_mean",

    "hallucination_prevalence",

    "hallucination_density",

    "compression_ratio_mean"
]

# =====================================================================
# 10. CHECK WHICH METRICS EXIST IN DATAFRAME
# =====================================================================

available_positive_metrics = [

    metric
    for metric in positive_metrics
    if metric in combined_df.columns
]

available_negative_metrics = [

    metric
    for metric in negative_metrics
    if metric in combined_df.columns
]

print("\nPositive metrics used for benchmarking:")
print(available_positive_metrics)

print("\nNegative metrics used for benchmarking:")
print(available_negative_metrics)

# Create working dataframe
benchmark_df = combined_df.copy()

# =====================================================================
# 11. NORMALIZE METRICS
# =====================================================================
# Metric normalization is required because:
#
# - Different metrics have different scales
# - Some metrics range from 0-1
# - Some metrics may have larger numerical ranges
#
# Min-Max normalization scales all metrics between:
#
# 0 and 1
#
# =====================================================================

scaler = MinMaxScaler()

all_metrics = (
    available_positive_metrics +
    available_negative_metrics
)

# Convert metrics to numeric values
for metric in all_metrics:

    benchmark_df[metric] = pd.to_numeric(
        benchmark_df[metric],
        errors="coerce"
    )

# Replace missing values with zero
benchmark_df[all_metrics] = benchmark_df[all_metrics].fillna(0)

# Perform normalization
normalized_values = scaler.fit_transform(
    benchmark_df[all_metrics]
)

# Store normalized metrics
for i, metric in enumerate(all_metrics):

    benchmark_df[metric + "_norm"] = normalized_values[:, i]

print("\nMetric normalization completed successfully")

# =====================================================================
# 12. INVERT NEGATIVE METRICS
# =====================================================================
# For negative metrics:
#
# Lower value = better performance
#
# Therefore:
#
# score = 1 - normalized_value
#
# =====================================================================

for metric in available_negative_metrics:

    benchmark_df[metric + "_score"] = (
        1 - benchmark_df[metric + "_norm"]
    )

# =====================================================================
# 13. POSITIVE METRICS SCORE
# =====================================================================
# For positive metrics:
#
# Higher value = better performance
#
# Therefore:
#
# score = normalized_value
#
# =====================================================================

for metric in available_positive_metrics:

    benchmark_df[metric + "_score"] = (
        benchmark_df[metric + "_norm"]
    )

print("\nMetric scoring completed successfully")

# =====================================================================
# 14. DEFINE WEIGHTS FOR COMPOSITE BENCHMARK SCORE
# =====================================================================
# Different metrics contribute differently to final evaluation.
#
# Higher weights are assigned to:
#
# - Semantic quality
# - Information retention
#
# Lower weights are assigned to:
#
# - Compression
# - Hallucination-related metrics
#
# =====================================================================

weights = {

    "semantic_similarity_mean_score": 0.25,

    "rougeL_f1_mean_score": 0.20,

    "irr_mean_score": 0.20,

    "top_k_skill_coverage_mean_score": 0.15,

    "global_skill_overlap_score": 0.10,

    "hallucination_rate_mean_score": 0.05,

    "hallucination_density_score": 0.03,

    "compression_ratio_mean_score": 0.02
}

# Keep only weights for available metrics
available_weights = {

    metric: weight

    for metric, weight in weights.items()

    if metric in benchmark_df.columns
}

print("\nWeights used in benchmark score calculation:")

for metric, weight in available_weights.items():

    print(metric, ":", weight)

# =====================================================================
# 15. CALCULATE COMPOSITE BENCHMARK SCORE
# =====================================================================
# Final benchmark score is calculated as:
#
# Weighted sum of all metric scores
#
# =====================================================================

benchmark_df["benchmark_score"] = 0

for metric, weight in available_weights.items():

    benchmark_df["benchmark_score"] += (

        benchmark_df[metric] * weight
    )

# Normalize benchmark score by total weight
total_weight = sum(available_weights.values())

benchmark_df["benchmark_score"] = (

    benchmark_df["benchmark_score"] / total_weight
)

print("\nComposite benchmark score calculated successfully")

# =====================================================================
# 16. OVERALL EXPERIMENT RANKING
# =====================================================================
# This ranking identifies:
#
# - Best overall experiment
# - Worst overall experiment
#
# across all:
#
# - prompts
# - models
# - RAG configurations
#
# =====================================================================

overall_ranking = benchmark_df.sort_values(

    by="benchmark_score",

    ascending=False
).reset_index(drop=True)

overall_ranking["overall_rank"] = (

    overall_ranking.index + 1
)

# Save ranking
overall_ranking.to_csv(

    OUTPUT_DIR / "overall_experiment_ranking.csv",

    index=False
)

print("\nOverall experiment ranking saved successfully")

print("\nTop ranked experiments:")

display(
    overall_ranking.head(10)
)

# =====================================================================
# 16A. SAVE OVERALL RANKING SEPARATELY FOR RAG AND NON-RAG
# =====================================================================
# Although the combined ranking is useful for comparing all experiments
# together, separate ranking files are also created for clearer analysis.
#
# This helps the examiner review:
#
# - Best experiments within the RAG configuration
# - Best experiments within the Non-RAG configuration
#
# =====================================================================

for rag_type in overall_ranking["rag_type"].unique():

    separate_ranking = overall_ranking[
        overall_ranking["rag_type"] == rag_type
    ].copy()

    separate_ranking = separate_ranking.sort_values(
        by="benchmark_score",
        ascending=False
    ).reset_index(drop=True)

    separate_ranking["rank_within_rag_type"] = (
        separate_ranking.index + 1
    )

    output_file_name = (
        f"overall_experiment_ranking_{rag_type}.csv"
    )

    separate_ranking.to_csv(
        OUTPUT_DIR / output_file_name,
        index=False
    )

    print(
        f"Saved separate overall experiment ranking file: {output_file_name}"
    )

# =====================================================================
# 17. PROMPTING TECHNIQUE RANKING
# =====================================================================
# This section ranks prompting techniques:
#
# - separately for each LLM
# - separately for RAG and Non-RAG
#
# =====================================================================

prompt_ranking = (

    benchmark_df

    .groupby(

        ["rag_type", "model_name", "prompt_type"],

        as_index=False
    )

    ["benchmark_score"]

    .mean()
)

prompt_ranking["prompt_rank"] = (

    prompt_ranking

    .groupby(["rag_type", "model_name"])

    ["benchmark_score"]

    .rank(

        ascending=False,

        method="dense"
    )

    .astype(int)
)

prompt_ranking = prompt_ranking.sort_values(

    ["rag_type", "model_name", "prompt_rank"]
)

# Save results
prompt_ranking.to_csv(

    OUTPUT_DIR / "prompt_ranking_by_model_and_rag_type.csv",

    index=False
)

print("\nPrompt ranking completed successfully")

display(prompt_ranking)

# =====================================================================
# 17A. SAVE PROMPT RANKING SEPARATELY FOR RAG AND NON-RAG
# =====================================================================

for rag_type in prompt_ranking["rag_type"].unique():

    separate_prompt_ranking = prompt_ranking[
        prompt_ranking["rag_type"] == rag_type
    ].copy()

    output_file_name = (
        f"prompt_ranking_by_model_{rag_type}.csv"
    )

    separate_prompt_ranking.to_csv(
        OUTPUT_DIR / output_file_name,
        index=False
    )

    print(
        f"Saved separate prompt ranking file: {output_file_name}"
    )


# =====================================================================
# 18. MODEL RANKING
# =====================================================================
# This section ranks LLMs:
#
# - separately for each prompting technique
# - separately for RAG and Non-RAG
#
# =====================================================================

model_ranking = (

    benchmark_df

    .groupby(

        ["rag_type", "prompt_type", "model_name"],

        as_index=False
    )

    ["benchmark_score"]

    .mean()
)

model_ranking["model_rank"] = (

    model_ranking

    .groupby(["rag_type", "prompt_type"])

    ["benchmark_score"]

    .rank(

        ascending=False,

        method="dense"
    )

    .astype(int)
)

model_ranking = model_ranking.sort_values(

    ["rag_type", "prompt_type", "model_rank"]
)

# Save results
model_ranking.to_csv(

    OUTPUT_DIR / "model_ranking_by_prompt_and_rag_type.csv",

    index=False
)

print("\nModel ranking completed successfully")

display(model_ranking)

# =====================================================================
# 18A. SAVE MODEL RANKING SEPARATELY FOR RAG AND NON-RAG
# =====================================================================

for rag_type in model_ranking["rag_type"].unique():

    separate_model_ranking = model_ranking[
        model_ranking["rag_type"] == rag_type
    ].copy()

    output_file_name = (
        f"model_ranking_by_prompt_{rag_type}.csv"
    )

    separate_model_ranking.to_csv(
        OUTPUT_DIR / output_file_name,
        index=False
    )

    print(
        f"Saved separate model ranking file: {output_file_name}"
    )


# =====================================================================
# 19. RAG VS NON-RAG COMPARISON
# =====================================================================
# This section compares:
#
# - RAG performance
# - Non-RAG performance
#
# for:
#
# - same prompt
# - same model
#
# =====================================================================

rag_pivot = benchmark_df.pivot_table(

    index=["prompt_type", "model_name"],

    columns="rag_type",

    values="benchmark_score",

    aggfunc="mean"
).reset_index()

# Calculate RAG gain
if "RAG" in rag_pivot.columns and "NON_RAG" in rag_pivot.columns:

    rag_pivot["rag_gain"] = (

        rag_pivot["RAG"] -

        rag_pivot["NON_RAG"]
    )

    rag_pivot["rag_gain_percentage"] = (

        rag_pivot["rag_gain"]

        / rag_pivot["NON_RAG"]
    ) * 100

# Sort by gain
rag_pivot = rag_pivot.sort_values(

    "rag_gain",

    ascending=False
)

# Save results
rag_pivot.to_csv(

    OUTPUT_DIR / "rag_vs_nonrag_comparison.csv",

    index=False
)

print("\nRAG vs Non-RAG comparison completed successfully")

display(rag_pivot)

# =====================================================================
# 20. BEST PROMPTING TECHNIQUE OVERALL
# =====================================================================

best_prompt_overall = (

    benchmark_df

    .groupby(

        ["rag_type", "prompt_type"],

        as_index=False
    )

    ["benchmark_score"]

    .mean()
)

best_prompt_overall["rank"] = (

    best_prompt_overall

    .groupby("rag_type")

    ["benchmark_score"]

    .rank(

        ascending=False,

        method="dense"
    )

    .astype(int)
)

best_prompt_overall = best_prompt_overall.sort_values(

    ["rag_type", "rank"]
)

best_prompt_overall.to_csv(

    OUTPUT_DIR / "best_prompt_overall.csv",

    index=False
)

print("\nBest prompting technique analysis completed")

display(best_prompt_overall)

# =====================================================================
# 21. BEST MODEL OVERALL
# =====================================================================

best_model_overall = (

    benchmark_df

    .groupby(

        ["rag_type", "model_name"],

        as_index=False
    )

    ["benchmark_score"]

    .mean()
)

best_model_overall["rank"] = (

    best_model_overall

    .groupby("rag_type")

    ["benchmark_score"]

    .rank(

        ascending=False,

        method="dense"
    )

    .astype(int)
)

best_model_overall = best_model_overall.sort_values(

    ["rag_type", "rank"]
)

best_model_overall.to_csv(

    OUTPUT_DIR / "best_model_overall.csv",

    index=False
)

print("\nBest model analysis completed")

display(best_model_overall)

# =====================================================================
# 22. GENERATE BAR CHARTS
# =====================================================================
# Charts are generated for:
#
# - Prompt rankings
# - Model rankings
#
# =====================================================================

for rag_type in best_prompt_overall["rag_type"].unique():

    subset = best_prompt_overall[
        best_prompt_overall["rag_type"] == rag_type
    ]

    plt.figure(figsize=(10, 5))

    plt.bar(

        subset["prompt_type"],

        subset["benchmark_score"]
    )

    plt.title(
        f"Prompting Technique Ranking - {rag_type}"
    )

    plt.xlabel("Prompting Technique")

    plt.ylabel("Benchmark Score")

    plt.xticks(rotation=45)

    plt.tight_layout()

    plt.savefig(

        OUTPUT_DIR /
        f"prompt_ranking_{rag_type}.png"
    )

    plt.show()

print("\nPrompt ranking charts generated successfully")

# =====================================================================
# 23. GENERATE MODEL RANKING CHARTS
# =====================================================================

for rag_type in best_model_overall["rag_type"].unique():

    subset = best_model_overall[
        best_model_overall["rag_type"] == rag_type
    ]

    plt.figure(figsize=(10, 5))

    plt.bar(

        subset["model_name"],

        subset["benchmark_score"]
    )

    plt.title(
        f"Model Ranking - {rag_type}"
    )

    plt.xlabel("Model")

    plt.ylabel("Benchmark Score")

    plt.xticks(rotation=45)

    plt.tight_layout()

    plt.savefig(

        OUTPUT_DIR /
        f"model_ranking_{rag_type}.png"
    )

    plt.show()

print("\nModel ranking charts generated successfully")

# =====================================================================
# 24. GENERATE RAG IMPROVEMENT CHART
# =====================================================================

if "rag_gain_percentage" in rag_pivot.columns:

    rag_pivot["experiment_label"] = (

        rag_pivot["prompt_type"] +

        " | " +

        rag_pivot["model_name"]
    )

    plt.figure(figsize=(14, 6))

    plt.bar(

        rag_pivot["experiment_label"],

        rag_pivot["rag_gain_percentage"]
    )

    plt.title(
        "RAG Improvement Percentage by Experiment"
    )

    plt.xlabel("Prompt and Model Combination")

    plt.ylabel("RAG Gain Percentage")

    plt.xticks(rotation=90)

    plt.tight_layout()

    plt.savefig(

        OUTPUT_DIR /
        "rag_gain_percentage_chart.png"
    )

    plt.show()

print("\nRAG improvement chart generated successfully")

# =====================================================================
# 25. DISPLAY FINAL OUTPUT SUMMARY
# =====================================================================

print("\n" + "=" * 70)
print("BENCHMARKING PIPELINE COMPLETED SUCCESSFULLY")
print("=" * 70)

print("\nGenerated benchmark CSV files:")

for file in OUTPUT_DIR.glob("*.csv"):

    print("-", file.name)

print("\nGenerated benchmark charts:")

for file in OUTPUT_DIR.glob("*.png"):

    print("-", file.name)

print("\nAll benchmark outputs are saved in:")
print(OUTPUT_DIR)
```
