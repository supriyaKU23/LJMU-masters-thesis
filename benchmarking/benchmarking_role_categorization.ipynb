# Benchmarking and Comparative Analysis of Prompt-Based Role Categorization Experiments

## Overview

This notebook performs benchmarking and comparative evaluation of prompt-based role categorization experiments conducted on the Djinni Job Descriptions Dataset.

The experiments were conducted using:

- Multiple prompting techniques
- Multiple Large Language Models (LLMs)
- Retrieval-Augmented Generation (RAG) and Non-RAG configurations

The objective of this benchmarking pipeline is to:

1. Evaluate role categorization performance across experiments
2. Rank prompting techniques for each LLM
3. Rank LLMs for each prompting technique
4. Compare RAG and Non-RAG performance
5. Identify the best-performing experimental configurations
6. Generate quantitative benchmark statistics and visualizations

```python
# =====================================================================
# BENCHMARKING PIPELINE FOR ROLE CATEGORIZATION EXPERIMENTS
# =====================================================================
# Purpose:
# This notebook benchmarks role categorization experiments conducted with:
#
# 1. Multiple prompting techniques
# 2. Multiple Large Language Models
# 3. RAG and Non-RAG configurations
#
# The notebook performs:
#
# - Loading RAG and Non-RAG final metric files
# - Cleaning prompt names
# - Metric normalization
# - Composite benchmark score calculation
# - Overall experiment ranking
# - Separate RAG and Non-RAG rankings
# - Prompt ranking
# - Model ranking
# - RAG vs Non-RAG comparison
# - Benchmark visualization generation
#
# =====================================================================

# =====================================================================
# 1. IMPORT REQUIRED LIBRARIES
# =====================================================================

import pandas as pd
import numpy as np

from pathlib import Path
from sklearn.preprocessing import MinMaxScaler

import matplotlib.pyplot as plt

# =====================================================================
# 2. MOUNT GOOGLE DRIVE
# =====================================================================

from google.colab import drive

drive.mount("/content/drive")

print("=" * 70)
print("ROLE CATEGORIZATION BENCHMARKING PIPELINE")
print("=" * 70)

print("\nGoogle Drive mounted successfully")

# =====================================================================
# 3. DEFINE PROJECT DIRECTORY
# =====================================================================
# Update this path if your project folder path is different.
# =====================================================================

PROJECT_DIR = Path(
    "/content/drive/MyDrive/Colab Notebooks/Masters/LJMU-masters-thesis"
)

print("\nProject directory:")
print(PROJECT_DIR)

# =====================================================================
# 4. DEFINE INPUT FILE PATHS
# =====================================================================
# These two files should contain final metrics from:
#
# - Role categorization without RAG
# - Role categorization with RAG
#
# =====================================================================

NON_RAG_FILE = (
    PROJECT_DIR
    / "outputs"
    / "final_comparison"
    / "role_categorization"
    / "non_rag"
    / "nonrag_rc_final_metrics_all_experiments.csv"
)

RAG_FILE = (
    PROJECT_DIR
    / "outputs"
    / "final_comparison"
    / "role_categorization"
    / "rag"
    / "rag_rc_final_metrics_all_experiments.csv"
)

print("\nInput files:")

print("\nNon-RAG file:")
print(NON_RAG_FILE)

print("\nRAG file:")
print(RAG_FILE)

# =====================================================================
# 5. CREATE OUTPUT DIRECTORY
# =====================================================================

OUTPUT_DIR = (
    PROJECT_DIR
    / "outputs"
    / "benchmarking"
    / "role_categorization"
)

OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

print("\nOutput directory created successfully:")
print(OUTPUT_DIR)

# =====================================================================
# 6. LOAD FINAL METRIC FILES
# =====================================================================

non_rag_df = pd.read_csv(NON_RAG_FILE)
rag_df = pd.read_csv(RAG_FILE)

print("\nFiles loaded successfully")

print("\nNon-RAG dataframe shape:", non_rag_df.shape)
print("RAG dataframe shape    :", rag_df.shape)

# =====================================================================
# 7. ADD RAG TYPE LABEL
# =====================================================================

non_rag_df["rag_type"] = "NON_RAG"
rag_df["rag_type"] = "RAG"

# =====================================================================
# 8. CLEAN PROMPT TYPE NAMES
# =====================================================================
# This standardizes prompt names by removing RAG-specific suffixes.
# Example:
#
# zero_shot_with_rag      -> zero_shot
# zero_shot_without_rag   -> zero_shot
#
# =====================================================================

rag_df["prompt_type"] = (
    rag_df["prompt_type"]
    .str.replace("_with_rag", "", regex=False)
    .str.replace("with_rag", "", regex=False)
)

non_rag_df["prompt_type"] = (
    non_rag_df["prompt_type"]
    .str.replace("_without_rag", "", regex=False)
    .str.replace("without_rag", "", regex=False)
)

print("\nPrompt types after cleaning")

print("\nNon-RAG prompt types:")
print(non_rag_df["prompt_type"].unique())

print("\nRAG prompt types:")
print(rag_df["prompt_type"].unique())

# =====================================================================
# 9. COMBINE RAG AND NON-RAG DATA
# =====================================================================

combined_df = pd.concat(
    [non_rag_df, rag_df],
    ignore_index=True,
    sort=False
)

print("\nCombined dataframe shape:", combined_df.shape)

combined_df.to_csv(
    OUTPUT_DIR / "combined_role_categorization_rag_nonrag_metrics.csv",
    index=False
)

print("\nCombined metrics file saved successfully")

# =====================================================================
# 10. DEFINE BENCHMARK METRICS
# =====================================================================
# For role categorization, classification metrics are used.
#
# Positive metrics:
# Higher value indicates better performance.
#
# Negative metrics:
# Lower value indicates better performance.
#
# =====================================================================

positive_metrics = [

    "accuracy",

    "precision_macro",

    "recall_macro",

    "f1_macro",

    "precision_weighted",

    "recall_weighted",

    "f1_weighted",

    "valid_label_rate"
]

negative_metrics = [

    "num_incorrect"
]

available_positive_metrics = [
    metric
    for metric in positive_metrics
    if metric in combined_df.columns
]

available_negative_metrics = [
    metric
    for metric in negative_metrics
    if metric in combined_df.columns
]

print("\nPositive metrics used:")
print(available_positive_metrics)

print("\nNegative metrics used:")
print(available_negative_metrics)

benchmark_df = combined_df.copy()

# =====================================================================
# 11. NORMALIZE METRICS
# =====================================================================
# Normalization is required because different metrics may have different
# numerical ranges.
#
# Min-Max scaling converts all metric values into the range 0 to 1.
# =====================================================================

scaler = MinMaxScaler()

all_metrics = available_positive_metrics + available_negative_metrics

for metric in all_metrics:

    benchmark_df[metric] = pd.to_numeric(
        benchmark_df[metric],
        errors="coerce"
    )

benchmark_df[all_metrics] = benchmark_df[all_metrics].fillna(0)

normalized_values = scaler.fit_transform(
    benchmark_df[all_metrics]
)

for i, metric in enumerate(all_metrics):

    benchmark_df[metric + "_norm"] = normalized_values[:, i]

print("\nMetric normalization completed successfully")

# =====================================================================
# 12. CREATE METRIC SCORES
# =====================================================================
# Positive metrics:
# Higher value is better, so normalized value is used directly.
#
# Negative metrics:
# Lower value is better, so inverted normalized value is used.
# =====================================================================

for metric in available_positive_metrics:

    benchmark_df[metric + "_score"] = (
        benchmark_df[metric + "_norm"]
    )

for metric in available_negative_metrics:

    benchmark_df[metric + "_score"] = (
        1 - benchmark_df[metric + "_norm"]
    )

print("\nMetric scoring completed successfully")

# =====================================================================
# 13. DEFINE WEIGHTS FOR ROLE CATEGORIZATION BENCHMARK SCORE
# =====================================================================
# Macro F1 is given the highest weight because the role categorization
# dataset may contain class imbalance.
#
# Accuracy is useful but not sufficient alone.
#
# Weighted F1 is also included because it reflects performance while
# considering class frequency.
#
# Valid label rate is included to reward outputs that correctly map to
# the fixed role taxonomy.
# =====================================================================

weights = {

    "f1_macro_score": 0.30,

    "accuracy_score": 0.20,

    "precision_macro_score": 0.15,

    "recall_macro_score": 0.15,

    "f1_weighted_score": 0.10,

    "valid_label_rate_score": 0.05,

    "num_incorrect_score": 0.05
}

available_weights = {
    metric: weight
    for metric, weight in weights.items()
    if metric in benchmark_df.columns
}

print("\nWeights used in benchmark score calculation:")

for metric, weight in available_weights.items():

    print(metric, ":", weight)

# =====================================================================
# 14. CALCULATE COMPOSITE BENCHMARK SCORE
# =====================================================================

benchmark_df["benchmark_score"] = 0

for metric, weight in available_weights.items():

    benchmark_df["benchmark_score"] += (
        benchmark_df[metric] * weight
    )

total_weight = sum(available_weights.values())

benchmark_df["benchmark_score"] = (
    benchmark_df["benchmark_score"] / total_weight
)

print("\nComposite benchmark score calculated successfully")

# =====================================================================
# 15. OVERALL EXPERIMENT RANKING
# =====================================================================

overall_ranking = benchmark_df.sort_values(
    by="benchmark_score",
    ascending=False
).reset_index(drop=True)

overall_ranking["overall_rank"] = overall_ranking.index + 1

overall_ranking.to_csv(
    OUTPUT_DIR / "overall_role_categorization_experiment_ranking.csv",
    index=False
)

print("\nOverall experiment ranking saved successfully")

display(overall_ranking.head(10))

# =====================================================================
# 16. SAVE OVERALL RANKING SEPARATELY FOR RAG AND NON-RAG
# =====================================================================

for rag_type in overall_ranking["rag_type"].unique():

    separate_ranking = overall_ranking[
        overall_ranking["rag_type"] == rag_type
    ].copy()

    separate_ranking = separate_ranking.sort_values(
        by="benchmark_score",
        ascending=False
    ).reset_index(drop=True)

    separate_ranking["rank_within_rag_type"] = (
        separate_ranking.index + 1
    )

    output_file_name = (
        f"overall_role_categorization_experiment_ranking_{rag_type}.csv"
    )

    separate_ranking.to_csv(
        OUTPUT_DIR / output_file_name,
        index=False
    )

    print("Saved:", output_file_name)

# =====================================================================
# 17. PROMPTING TECHNIQUE RANKING
# =====================================================================
# This ranks prompting techniques separately for:
#
# - each LLM
# - RAG configuration
# - Non-RAG configuration
# =====================================================================

prompt_ranking = (
    benchmark_df
    .groupby(
        ["rag_type", "model_name", "prompt_type"],
        as_index=False
    )
    ["benchmark_score"]
    .mean()
)

prompt_ranking["prompt_rank"] = (
    prompt_ranking
    .groupby(["rag_type", "model_name"])
    ["benchmark_score"]
    .rank(
        ascending=False,
        method="dense"
    )
    .astype(int)
)

prompt_ranking = prompt_ranking.sort_values(
    ["rag_type", "model_name", "prompt_rank"]
)

prompt_ranking.to_csv(
    OUTPUT_DIR / "prompt_ranking_by_model_and_rag_type.csv",
    index=False
)

print("\nPrompt ranking completed successfully")

display(prompt_ranking)

# =====================================================================
# 18. SAVE PROMPT RANKING SEPARATELY FOR RAG AND NON-RAG
# =====================================================================

for rag_type in prompt_ranking["rag_type"].unique():

    separate_prompt_ranking = prompt_ranking[
        prompt_ranking["rag_type"] == rag_type
    ].copy()

    output_file_name = (
        f"prompt_ranking_by_model_{rag_type}.csv"
    )

    separate_prompt_ranking.to_csv(
        OUTPUT_DIR / output_file_name,
        index=False
    )

    print("Saved:", output_file_name)

# =====================================================================
# 19. MODEL RANKING
# =====================================================================
# This ranks LLMs separately for:
#
# - each prompting technique
# - RAG configuration
# - Non-RAG configuration
# =====================================================================

model_ranking = (
    benchmark_df
    .groupby(
        ["rag_type", "prompt_type", "model_name"],
        as_index=False
    )
    ["benchmark_score"]
    .mean()
)

model_ranking["model_rank"] = (
    model_ranking
    .groupby(["rag_type", "prompt_type"])
    ["benchmark_score"]
    .rank(
        ascending=False,
        method="dense"
    )
    .astype(int)
)

model_ranking = model_ranking.sort_values(
    ["rag_type", "prompt_type", "model_rank"]
)

model_ranking.to_csv(
    OUTPUT_DIR / "model_ranking_by_prompt_and_rag_type.csv",
    index=False
)

print("\nModel ranking completed successfully")

display(model_ranking)

# =====================================================================
# 20. SAVE MODEL RANKING SEPARATELY FOR RAG AND NON-RAG
# =====================================================================

for rag_type in model_ranking["rag_type"].unique():

    separate_model_ranking = model_ranking[
        model_ranking["rag_type"] == rag_type
    ].copy()

    output_file_name = (
        f"model_ranking_by_prompt_{rag_type}.csv"
    )

    separate_model_ranking.to_csv(
        OUTPUT_DIR / output_file_name,
        index=False
    )

    print("Saved:", output_file_name)

# =====================================================================
# 21. RAG VS NON-RAG COMPARISON
# =====================================================================
# This compares the same prompt and same model under:
#
# - RAG
# - Non-RAG
#
# =====================================================================

rag_pivot = benchmark_df.pivot_table(
    index=["prompt_type", "model_name"],
    columns="rag_type",
    values="benchmark_score",
    aggfunc="mean"
).reset_index()

if "RAG" in rag_pivot.columns and "NON_RAG" in rag_pivot.columns:

    rag_pivot["rag_gain"] = (
        rag_pivot["RAG"] - rag_pivot["NON_RAG"]
    )

    rag_pivot["rag_gain_percentage"] = (
        rag_pivot["rag_gain"] / rag_pivot["NON_RAG"]
    ) * 100

rag_pivot = rag_pivot.sort_values(
    "rag_gain",
    ascending=False
)

rag_pivot.to_csv(
    OUTPUT_DIR / "rag_vs_nonrag_role_categorization_comparison.csv",
    index=False
)

print("\nRAG vs Non-RAG comparison completed successfully")

display(rag_pivot)

# =====================================================================
# 22. BEST PROMPTING TECHNIQUE OVERALL
# =====================================================================

best_prompt_overall = (
    benchmark_df
    .groupby(
        ["rag_type", "prompt_type"],
        as_index=False
    )
    ["benchmark_score"]
    .mean()
)

best_prompt_overall["rank"] = (
    best_prompt_overall
    .groupby("rag_type")
    ["benchmark_score"]
    .rank(
        ascending=False,
        method="dense"
    )
    .astype(int)
)

best_prompt_overall = best_prompt_overall.sort_values(
    ["rag_type", "rank"]
)

best_prompt_overall.to_csv(
    OUTPUT_DIR / "best_prompt_overall_role_categorization.csv",
    index=False
)

print("\nBest prompting technique analysis completed")

display(best_prompt_overall)

# =====================================================================
# 23. BEST MODEL OVERALL
# =====================================================================

best_model_overall = (
    benchmark_df
    .groupby(
        ["rag_type", "model_name"],
        as_index=False
    )
    ["benchmark_score"]
    .mean()
)

best_model_overall["rank"] = (
    best_model_overall
    .groupby("rag_type")
    ["benchmark_score"]
    .rank(
        ascending=False,
        method="dense"
    )
    .astype(int)
)

best_model_overall = best_model_overall.sort_values(
    ["rag_type", "rank"]
)

best_model_overall.to_csv(
    OUTPUT_DIR / "best_model_overall_role_categorization.csv",
    index=False
)

print("\nBest model analysis completed")

display(best_model_overall)

# =====================================================================
# 24. GENERATE PROMPT RANKING CHARTS
# =====================================================================

for rag_type in best_prompt_overall["rag_type"].unique():

    subset = best_prompt_overall[
        best_prompt_overall["rag_type"] == rag_type
    ]

    plt.figure(figsize=(10, 5))

    plt.bar(
        subset["prompt_type"],
        subset["benchmark_score"]
    )

    plt.title(
        f"Prompting Technique Ranking for Role Categorization - {rag_type}"
    )

    plt.xlabel("Prompting Technique")

    plt.ylabel("Benchmark Score")

    plt.xticks(rotation=45)

    plt.tight_layout()

    plt.savefig(
        OUTPUT_DIR / f"prompt_ranking_role_categorization_{rag_type}.png"
    )

    plt.show()

print("\nPrompt ranking charts generated successfully")

# =====================================================================
# 25. GENERATE MODEL RANKING CHARTS
# =====================================================================

for rag_type in best_model_overall["rag_type"].unique():

    subset = best_model_overall[
        best_model_overall["rag_type"] == rag_type
    ]

    plt.figure(figsize=(10, 5))

    plt.bar(
        subset["model_name"],
        subset["benchmark_score"]
    )

    plt.title(
        f"Model Ranking for Role Categorization - {rag_type}"
    )

    plt.xlabel("Model")

    plt.ylabel("Benchmark Score")

    plt.xticks(rotation=45)

    plt.tight_layout()

    plt.savefig(
        OUTPUT_DIR / f"model_ranking_role_categorization_{rag_type}.png"
    )

    plt.show()

print("\nModel ranking charts generated successfully")

# =====================================================================
# 26. GENERATE RAG IMPROVEMENT CHART
# =====================================================================

if "rag_gain_percentage" in rag_pivot.columns:

    rag_pivot["experiment_label"] = (
        rag_pivot["prompt_type"] + " | " + rag_pivot["model_name"]
    )

    plt.figure(figsize=(14, 6))

    plt.bar(
        rag_pivot["experiment_label"],
        rag_pivot["rag_gain_percentage"]
    )

    plt.title(
        "RAG Improvement Percentage for Role Categorization"
    )

    plt.xlabel("Prompt and Model Combination")

    plt.ylabel("RAG Gain Percentage")

    plt.xticks(rotation=90)

    plt.tight_layout()

    plt.savefig(
        OUTPUT_DIR / "rag_gain_percentage_role_categorization.png"
    )

    plt.show()

print("\nRAG improvement chart generated successfully")

# =====================================================================
# 27. DISPLAY FINAL OUTPUT SUMMARY
# =====================================================================

print("\n" + "=" * 70)
print("ROLE CATEGORIZATION BENCHMARKING COMPLETED SUCCESSFULLY")
print("=" * 70)

print("\nGenerated CSV files:")

for file in OUTPUT_DIR.glob("*.csv"):

    print("-", file.name)

print("\nGenerated chart files:")

for file in OUTPUT_DIR.glob("*.png"):

    print("-", file.name)

print("\nAll outputs are saved in:")
print(OUTPUT_DIR)
```
